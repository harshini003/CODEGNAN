# We will generate some data and will understand how it is Underfitted,how it is overfitted and how it can be Ideal Fit(Balanced Fit)

```python
#import the libraries
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#we will generate some random data and check with it now let us use uniform function
#to generate synthetic data
n_points = 20 #based on these points let us apply uniform function
x = np.random.uniform(0,11,n_points)
#x
#now you can sort the values
x = np.sort(x)
x
#x.shape
```

```python
#defining dependent variable
y = (-x+2)*(x-9) + np.random.normal(0,3,n_points) #Probabilty Distribution 
y
#y.shape
```

This looks like the shape of a parabola, as expected for a quadratic transformation of x. However we can see the noise in the fact that not all the points appear that they would lie perfectly on the parabola.

In our synthetic example, we know the data generating process: the response variable y is a quadratic transformation of the feature x. In general, when building machine learning models, the data generating process is not known. Instead, several candidate features are proposed, a model is proposed, and an exploration is made of how well these features and this model can explain the data.

In this case, we would likely plot the data, observe the apparent quadratic relationship, and use a quadratic model. However, for the sake of illustration of an underfit model, what does a linear model for these data look like?

# However for the sake of illustration of an underfit model,what does a linear model for these data look like?

```python
#based on above defined dependent and independent variables we will fit the data to the model
#first let's display the data 
#rcParams is runtime configuration parametes (dpi is dots per inches)
lin_fit = np.polyfit(x,y,1) #ordinary least squares using Polyfit
print(lin_fit)
#now based on the values generated 
plt.scatter(x,y,label = "Data Points")
plt.plot(x,np.polyval(lin_fit,x),color="red",label="Underfitted Data")
plt.legend()
plt.show()
```

```python
#now we will assume degree as higher 
```

```python
high_fit=np.polyfit(x,y,15)#odinary least square using polyfit
#print(high_fit)
plt.scatter(x,y,label = "Data Points")
plt.plot(x,np.polyval(high_fit,x),color="red",label="overerfitted Data")
plt.legend()
plt.show()
```

```python
import matplotlib as mpl
cmap=mpl.c
plt.scatter(x,y,label="Data Points")
plt.plot(x,np.polyval(high_fit,x)color="Black",label="Over Flow Data")
plt.legend()
plt.show()
```

```python
import matplotlib as mpl
cmap=mpl.cm.get_cmap('tab10')
cmap
plt.scatter(x,y,label="Datapoints",color=cmap(0))
plt.plot(x,np.polyval(lin_fit,x),label="Underfitted Data",color=cmap(2))
plt.plot(x,np.polyval(high_fit,x),label="Overfitted Data",color=cmap(4))
```

```python
#As we have our dependent variable with degree 2 we will assume proper Idealfit
#would be by taking value as 2 in polyfit() as below 
```

```python
plt.scatter(x,y,label="Data",color=cmap(0))
plt.plot(x,np.polyval(lin_fit,x),label="Underfitted Data",color=cmap(2))
plt.plot(x,np.polyval(np.polyfit(x,y,2),x),label="Ideal Fit",color=cmap(3))
plt.plot(x,np.polyval(high_fit,x),label="Underfitted Data",color=cmap(4))
plt.legend()
plt.show()
```
