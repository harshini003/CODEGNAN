```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#we will generate some random data-->random module
x=np.random.rand(5,2)#(5)# for 1 degree only linera generates
x
```

```python
#based on abovedata we will create set of points along with target 
#we wanted to generate some non-linear data
m=100#datapoints
x=6*np.random.rand(m,1)-1
x.shape
```

```python
#some using above input we will create a 2degree equation(ax^2+bx+c)
y=-0.5*x**2+x+np.random.rand(m,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(x,y)
plt.title("sample Representation")
plt.show()
```

```python
#by above graph we can say a atraight line will not fit the data pattern
#model fitting part
model=LinearRegression()
model.fit(x,y)
```

```python
print(dir(model))#will return methods available in LinearRegression
```

```python
print(model.coef_)
print(model.intercept_)#you can check it in your y varaiable
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
predict=model.predict(x)
#predict
```

```python
print("Mean Squared Error is",mean_squared_error(y,predict))
print("R-Squared value is",r2_score(y,predict))
```

```python
#once we evaluted the metrics we can visualize whether the regression model has deliverd best fit line or not
plt.scatter(x,y,color="yellow",label="Data Points")
plt.plot(x,predict,color="blue",label="Regression Line")
#always plot original data and test data
plt.legend()to display the label
plt.show()
```
